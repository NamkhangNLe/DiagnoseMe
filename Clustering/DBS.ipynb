```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture as GMM
from sklearn.cluster import DBSCAN as DBS
from sklearn import metrics
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.decomposition import PCA
from matplotlib.patches import Ellipse
```

```python
data = pd.read_csv('CleanedData/disease_vector.csv', header=0, index_col=0)
print(data)
```

```python
dbs = DBS(eps=2.4, min_samples=1)
labels = dbs.fit_predict(data)
print(labels)

diseases = []

for index, row in data.iterrows():
    diseases.append(index)

for i in range(len(diseases)):
    print(str(diseases[i]) + " " + str(labels[i]))
```

```python
n = np.zeros((1, 172))
for index, row in data.iterrows():
    if("epatitis " in index):
        n = np.add(row.values, n)

n = n/4

data = data[~data.index.str.contains("epatitis ")]
# data['Hepatitis A/B/C/D'] = n

new_row = pd.DataFrame(n, columns=data.columns, index=['Hepatitis A/B/C/D'])
data2 = pd.concat([data, new_row])

data2



```

```python
dbs = DBS(eps=1.95, min_samples=1)
labels = dbs.fit_predict(data2)
print(labels)

diseases = []

for index, row in data.iterrows():
    diseases.append(index)

for i in range(len(diseases)):
    print(str(diseases[i]) + " " + str(labels[i]))
```
